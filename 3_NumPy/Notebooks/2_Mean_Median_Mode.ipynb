# NumPy: Mean, Median, Mode

```python
# pip install numpy scipy matplotlib seaborn

import numpy as np
from scipy import stats

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## Mean, Median

```python

grades = np.random.randint(65, 101, size=25) # Generate 25 random grades between 65 and 100
print(f"Grades: {type(grades)}")
print(f"Grades: {grades}")

print("Mean:", np.mean(grades)) # Mean is average

sorted_grades = np.sort(grades)
print("Sorted Grades:", sorted_grades)
print("Median:", np.median(sorted_grades)) # Median is middle value
```

## Mode (SciPy)

```python

grades_mode = stats.mode(grades) # Uses scipy.stats to calculate the mode
print("Mode:", grades_mode.mode) # Mode is most frequently occurring value
```

## Random Distribution

```python
# Generate 1000 random salaries from a normal distribution with mean 25000 and std deviation 15000
salaries = np.random.normal(25000, 15000, 1000)
print(f"Salaries: {type(salaries)}")
print(f"Salaries: {len(salaries)}")

print("Mean:", f"${np.mean(salaries):,.2f}")
print("Median:", f"${np.median(salaries):,.2f}")
```

```python
# Plot Histogram of Salaries
sns.histplot(salaries, bins=50, kde=True, edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salaries')
plt.show()
```
